```python
import pandas as pd

df = pd.read_csv('FT1000.csv')
plt.figure(figsize=(12,6))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
top20_countries = df['Country'].value_counts().nlargest(20)
plt.figure(figsize=(12,6))
sns.barplot(x=top20_countries.index, y=top20_countries.values)
plt.title('Top 20 countries with the highest number of fastest growing companies')
plt.xlabel('Country')
plt.ylabel('Number of companies')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.scatterplot(x='CAGR', y='Revenue2020', data=df)
plt.title('Relationship between Growth Rate and Revenue')
plt.xlabel('Growth Rate')
plt.ylabel('Revenue Growth')
plt.show()
```

```python
sns.histplot(df['CAGR'], kde=True)
plt.title('Distribution of Growth Rates')
plt.xlabel('Growth Rate')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x='Sector', y='CAGR', data=df)
plt.title('Distribution of Growth Rates by Industry')
plt.xlabel('Industry')
plt.ylabel('Growth Rate')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.scatterplot(x='CAGR', y='Sector', data=df, hue='Country', palette='bright')
plt.title('Relationship between Growth Rate and Industry, Colored by Country')
plt.xlabel('Growth Rate')
plt.ylabel('Industry')
plt.show()
```

```python
sns.violinplot(x='Sector', y='CAGR', data=df, inner='quartile')
plt.title('Distribution of Growth Rates by Industry')
plt.xlabel('Industry')
plt.ylabel('Growth Rate')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.scatterplot(x='FoundingYear', y='CAGR', data=df)
plt.title('Relationship between Growth Rate and Company Age')
plt.xlabel('Company Age')
plt.ylabel('Growth Rate')
plt.show()
```

```python
sns.boxplot(x='Sector', y='Employees2020', data=df)
plt.title('Distribution of Employees by Industry')
plt.xlabel('Industry')
plt.ylabel('Number of Employees')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.scatterplot(x='Employees2020', y='Revenue2020', data=df)
plt.title('Relationship between Employees and Revenue')
plt.xlabel('Number of Employees')
plt.ylabel('Revenue')
plt.show()
```

```python
sns.boxplot(x='Country', y='Employees2020', data=df, palette='bright')
plt.title('Distribution of Employees by Country')
plt.xlabel('Country')
plt.ylabel('Number of Employees')
plt.xticks(rotation=90)
plt.show()
```

```python
cols = ['Employees2020', 'Revenue2020', 'CAGR', 'Country', 'Sector']
df = df[cols]

# Creating pair plot
sns.pairplot(df, hue='Country', palette='bright')
```
